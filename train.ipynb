# Definiendo el bucket y el rol

````python
import sagemaker

sagemaker_session = sagemaker.Session()

bucket = sagemaker_session.default_bucket()
prefix = 'sagemaker/example-requirements'

role = sagemaker.get_execution_role()
````

# Subiendo el contenido de la carpeta ```./data``` al bucket

````python
inputs = sagemaker_session.upload_data(path='./data', bucket=bucket, key_prefix=prefix)
````

# Definiendo el job
Se usará la imagen preparada para utilizar PyTorch en una instancia del tipo ml.m5.large. Pero el código (```./code/train.py```) que se va a ejecutar en la instancia necesita la biblioteca ```fastext```. Para que se instale esta biblioteca en dicha instancia se crea un archivo ```requirements.txt``` dentro del ```source_dir``` especificado. Como el ```source_dir``` es ```./code```, el archivo ```requirements.txt``` debe estar en ```./code/requirements.txt```. La existencia de este archivo dentro del ```source_dir``` hará que se ejecute el comando ```python -m pip install . -r requirements.txt``` automáticamente cuando se cree la instancia. Esto se puede apreciar al hacer 'fit' y ver cómo se ejecuta dicha línea automáticamente.

````python
from sagemaker.pytorch import PyTorch

estimator = PyTorch(entry_point='train.py',
                    role=role,
                    framework_version='1.4.0',
                    train_instance_count=1,
                    train_instance_type='ml.m5.large',
                    source_dir='./code'
                    )
````

````python
estimator.fit({'training': inputs})
````
